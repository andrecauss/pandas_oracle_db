```python
import pandas as pd
from connect_oracle import criar_conexao

conn = criar_conexao()
cursor = conn.cursor()
```

```python
cursor.execute("""
                    SELECT
                        TO_CHAR(o.order_date, 'YYYY-MM') AS month,
                        COUNT(o.order_id) as orders_quantity,
                        SUM(o.quantity) AS total_quantity,
                        SUM(o.sales) AS total_sales,
                        SUM(o.profit) AS total_profit,
                        SUM(o.discount) AS total_discount
                    FROM orders o
                    GROUP BY TO_CHAR(o.order_date, 'YYYY-MM')
                    HAVING SUM(o.quantity) > 200 and SUM(o.profit)>1000
                    ORDER BY MONTH ASC
               """)
```

```python
queryReturn = cursor.fetchall()
query = pd.DataFrame(columns=[i[0] for i in cursor.description], data=queryReturn)
```
